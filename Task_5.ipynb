```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data from the CSV file
accident_data = pd.read_csv("RTA Dataset.csv")
```

```python
# Display basic information about the dataset
print(accident_data.info())

```

```python
accident_data.head()
```

```python
# Display summary statistics for numerical columns
print(accident_data.describe())
```

```python
# Explore the distribution of accidents across different age
plt.figure(figsize=(12, 6))
sns.countplot(x="Age_band_of_driver", data=accident_data, order=accident_data['Age_band_of_driver'].value_counts().index)
plt.title("Distribution of Accidents Across Age")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Explore the distribution of accidents based on weather conditions
plt.figure(figsize=(12, 6))
sns.countplot(x="Weather_conditions", data=accident_data, order=accident_data['Weather_conditions'].value_counts().index)
plt.title("Distribution of Accidents Based on Weather Conditions")
plt.show()
```

```python
# Explore the distribution of accidents based on road conditions
plt.figure(figsize=(12, 6))
sns.countplot(x="Road_surface_conditions", data=accident_data, order=accident_data['Road_surface_conditions'].value_counts().index)
plt.title("Distribution of Accidents Based on Road Conditions")
plt.show()
```

```python
# Explore the distribution of accidents based on the timing of the day
plt.figure(figsize=(12, 6))
sns.countplot(x="Time", data=accident_data, order=accident_data['Time'].value_counts().index)
plt.title("Distribution of Accidents Based on Timing of the Day")
plt.show()
```

```python
# Explore the reasons for accidents
plt.figure(figsize=(12, 6))
sns.countplot(x="Cause_of_accident", data=accident_data, order=accident_data['Cause_of_accident'].value_counts().index)
plt.title("Distribution of Accidents Based on Reasons")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn plots
sns.set(style="whitegrid")

# Explore patterns related to road conditions
plt.figure(figsize=(12, 6))
sns.countplot(x="Road_surface_type", hue="Day_of_week", data=accident_data, order=accident_data['Road_surface_type'].value_counts().index)
plt.title("Accidents by Road Conditions and  Day_of_week")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Set the style for seaborn plots
sns.set(style="whitegrid")

# Visualize accident hotspots (States with higher accident frequencies)
plt.figure(figsize=(14, 8))
state_accidents = accident_data['Type_of_vehicle'].value_counts()
sns.barplot(x=state_accidents.index, y=state_accidents.values, palette="viridis")
plt.title("Accident Hotspots by Type_of_vehicle")
plt.xlabel("Type_of_vehicle")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Visualize contributing factors (Reasons for accidents)
plt.figure(figsize=(12, 6))
reasons_accidents = accident_data['Cause_of_accident'].value_counts()
sns.barplot(x=reasons_accidents.index, y=reasons_accidents.values, palette="muted")
plt.title("Contributing Factors for Accidents")
plt.xlabel("Cause_of_accident")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn plots
sns.set(style="whitegrid")

# Group the data by state and timing, and calculate the total number of deaths
state_timing_deaths = accident_data.groupby(['Type_of_vehicle', 'Sex_of_driver'])['Number_of_casualties'].sum().reset_index()

# Pivot the table to get a format suitable for plotting
state_timing_deaths_pivot = state_timing_deaths.pivot(index='Type_of_vehicle', columns='Sex_of_driver', values='Number_of_casualties').fillna(0)

# Create a bar plot for state-wise total death timing ratio
plt.figure(figsize=(40, 8))
state_timing_deaths_pivot.plot(kind='bar', stacked=True, colormap="viridis")
plt.title("Total Death Timing Ratio")
plt.xlabel("Type_of_vehicle")
plt.ylabel("Total Deaths")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Timing")
plt.show()
```
